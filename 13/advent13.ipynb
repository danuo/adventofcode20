```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

import re

with open('input') as file:
    lines = file.read().splitlines()
    
t_depart = int(lines[0])
timetable = lines[1]

bus_ids1 = re.findall(r'[0-9]+', timetable)
bus_ids1 = list(map(int, bus_ids1))

t_last_bus = [t_depart % bus_id for bus_id in bus_ids1]

t_next_bus = []
for t_last, bus_id in zip(t_last_bus, bus_ids1):
    t_next_bus.append(bus_id-t_last)
    
argmin, min_value = min(enumerate(t_next_bus), key=lambda x: x[1])

print('result 1: ', bus_ids1[argmin]*min_value)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

from functools import reduce

def extended_gcd(num1, num2):
	if num1 == 0:
		return (num2, 0, 1)
	else:
		gcd, x, y = extended_gcd(num2 % num1, num1)
		return (gcd, y - (num2//num1) * x, x)

def sync_reduce(input1, input2):
    # 
    # this function syncs two periodical systems A and B, that start with an
    # offset respectively:
    #
    # A: |<---- offset1 ---->|<----- x*step1----->|<---- 1*combined_step ---->|
    # B: |<-- offset2 -->|<--------- y*step2----->|<---- 1*combined_step ---->|
    #                                             ^              ^
    #                                      combined_offset      lcm
    #
    # the systems A and B stay in sync, when the Least Common Multiple (lcm)
    # is added
    #
    # first, we sync systems A and B to system AB. Then, we sync systems AB 
    # and C etc. until all systems are in sync.
    #
    # A,B -> AB
    # AB,C -> ABC
    # ABC,D -> ABCD ...
    #    
    
    step1, offset1 = input1
    step2, offset2 = input2
    offset_diff = offset2 - offset1
    
    # lcm is step1*step2, because both are prime
    lcm = step1 * step2
    gcd, x, y = extended_gcd(step1, step2)
    combined_offset = (offset1 + x * offset_diff * step1) % lcm
    return (lcm, combined_offset)

bus_ids2 = re.findall(r'[0-9]+|x', timetable)
bus_ids2 = map(lambda x:x.replace('x', '0'), bus_ids2)
bus_ids2 = zip(map(int, bus_ids2), range(0,-len(timetable),-1))
bus_ids2 = filter(lambda x: x[0] > 0, bus_ids2)
bus_ids2 = list(bus_ids2)

print('bus schedule (line_id, offset): ', bus_ids2)

result = reduce(sync_reduce, bus_ids2)

print('result 2: ', result[1])
#247086664214628
```
