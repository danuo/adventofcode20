```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

with open('input10') as file:
    lines = file.read().splitlines()

adapter_set = set(map(int, lines))
counter = [0] * 4
for _ in range(len(adapter_set)):
    i = 1
    while counter[0]+i not in adapter_set:
        i += 1
    counter[0] += i
    counter[i] += 1

print('result 1: ', counter[1]*(counter[3]+1))
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

from functools import lru_cache

permutation_list = []
for adapter in adapter_set | {0}: # with zero
    permutation_counter = 0
    for i in range(1,4):
        if adapter+i in adapter_set:
            permutation_counter += 1
    permutation_list.append(permutation_counter)

@lru_cache # input of eval_func needs to be hashable
def eval_func(k):
    section = permutation_list[k:]
    if len(section) == 1: return 1
    return sum([eval_func(k+i) for i in range(1,section[0]+1)])

print('result 2: ', eval_func(0))
```

```python
# ─── CHALLENGE 2 WITHOUT LRUCACHE ───────────────────────────────────────────────
# this solution is way more efficient and therefore can be executed quickly
# without utilizing function caching 

plist_rev = permutation_list.copy()
plist_rev.reverse()

outer_counter = [1] * len(plist_rev)
for i in range(1, len(plist_rev)):
    inner_counter = 0
    for j in range(1, plist_rev[i]+1):
        inner_counter += outer_counter[i-j]
    outer_counter[i] = inner_counter

print('result 2:', outer_counter[-1])
```
