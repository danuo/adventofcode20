```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────
import numpy as np
import re

with open('input04') as file:
    lines = file.read() #.splitlines()

# get each individual passport as list item (seperated by two line breaks)
passport_data_list = lines.split("\n\n")
passport_dict_list = []
for item in passport_data_list:
    item_dict = dict()
    # get each key from passport (seperated by white space)

    passport_strings = item.split()
    for string in passport_strings:
        key, value = string.split(':')
        item_dict.update({key: value})
    passport_dict_list.append(item_dict)

counter = 0
passport_dict_list_copy = passport_dict_list.copy()
for item in passport_dict_list_copy:
    # remove cid, not needed
    item.pop('cid', None)
    if len(item) == 7:
        counter += 1

print('result 1:', counter)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────
passport_dict_list_copy = passport_dict_list.copy()
counter = 0
for item in passport_dict_list_copy:
    # remove cid, not needed
    item.pop('cid', None)
    
    # check if all relevant keys are present
    if len(item) == 7: 
        
        # check if height is valid
        height_valid = False
        if item['hgt'].endswith('cm'):
            if 150 <= int(item['hgt'][:-2]) <= 193:
                height_valid = True
        if item['hgt'].endswith('in'):
            if 59 <= int(item['hgt'][:-2]) <= 76:
                height_valid = True

        # check all other conditions
        if all([
                1920 <= int(item['byr']) <= 2002,
                2010 <= int(item['iyr']) <= 2020,
                2020 <= int(item['eyr']) <= 2030,
                height_valid,
                bool(re.match("^#[a-f,0-9]{6}$", item['hcl'])),
                item['ecl'] in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth'],
                bool(re.match("^[0-9]{9}$", item['pid'])),
                ]):
            counter += 1

print('result 2:', counter)
```
