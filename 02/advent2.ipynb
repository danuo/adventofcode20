```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────
import re

with open('input') as file:
    lines = file.read().splitlines()

list_let, list_num_a, list_num_b, list_pas = [], [], [], []
for line in lines:
    result = re.split("\\W", line)
    result = list(filter(None, result)) # remove '' strings
    number_a, number_b, letter, password = result
    list_let.append(letter)
    list_num_a.append(int(number_a))
    list_num_b.append(int(number_b))
    list_pas.append(password)

counter = 0
for password, letter, num_a, num_b in zip(list_pas, list_let, list_num_a, list_num_b):
    letter_count = [char for char in password].count(letter)
    if num_a <= letter_count <= num_b:
        counter += 1
        
print('result 1:', counter)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

counter = 0
for password, letter, num_a, num_b in zip(list_pas, list_let, list_num_a, list_num_b):
    occurance_counter = 0
    if password[num_a - 1] == letter:
        occurance_counter += 1
    if password[num_b - 1] == letter:
        occurance_counter += 1
    if occurance_counter == 1:
        counter += 1
        
print('result 2:', counter)
```
