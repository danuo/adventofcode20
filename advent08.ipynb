```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

from functools import reduce

with open('input08') as file:
    lines = file.read().splitlines()

opp_list, val_list = [], []
for line in lines:
    opp, val = line.split(' ')
    opp_list.append(opp)
    val_list.append(int(val))
n = len(lines)

def run_code(opp_list, val_list):
    i, counter, exec_counter = 0, 0, [0] * n
    while True:
        # Challenge 1 termination: Terminate before 'acc' execution
        if exec_counter[i] == 1:
            return counter, i
        exec_counter[i] += 1
        
        if opp_list[i] == 'acc':
            counter += val_list[i]

        # Challenge 2 termination: Terminate after 'acc' execution
        if i == n-1:
            return counter, i

        if opp_list[i] == 'jmp':
            i += val_list[i]
            continue
        i += 1

print('result 1: ', run_code(opp_list, val_list)[0])
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

for j in range(n):
    if opp_list[j] == 'acc':
        continue
    opp_list_new = opp_list.copy()
    replacements = (('jmp','tmp'), ('nop','jmp'), ('tmp','nop'))
    opp_list_new[j] = reduce(lambda x, y : x.replace(*y), replacements, opp_list[j])
    counter, i = run_code(opp_list_new, val_list)
    if i == n - 1:
        break

print('result 2: ', counter)
```
