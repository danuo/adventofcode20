```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

inputs = [12,20,0,6,1,17,7]

counter = dict()

def check_number(t_now, number):
    if number in counter:
        t_last = counter[number]
        speak = t_now-t_last
    else:
        speak = 0
    counter.update({number: t_now})
    return speak


for t, number in enumerate(inputs, start=1):
    speak = check_number(t, number)

for t in range(1+len(inputs), 2020):
    speak = check_number(t, speak)
    
print('result 1: ', speak)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

for t in range(2020, 30000000):
    speak = check_number(t, speak)

print('result 2: ', speak)
```
