```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

import numpy as np
input_numbers = np.loadtxt('input', dtype=np.int)
matrix = input_numbers.reshape((-1,1))
matrix = np.repeat(matrix, input_numbers.size, axis=1)
matrix_summed = matrix + matrix.T
num_index = np.argwhere(matrix_summed == 2020)[0]
number1 = input_numbers[num_index[0]]
number2 = input_numbers[num_index[1]]
print('result 1:', number1 * number2)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

matrix = input_numbers.reshape((-1,1,1))
matrix = np.repeat(matrix, input_numbers.size, axis=1)
matrix = np.repeat(matrix, input_numbers.size, axis=2)
matrix_summed = np.transpose(matrix, axes=(0,1,2))\
              + np.transpose(matrix, axes=(1,0,2))\
              + np.transpose(matrix, axes=(2,1,0))  
num_index = np.argwhere(matrix_summed == 2020)[0]
number1 = input_numbers[num_index[0]]
number2 = input_numbers[num_index[1]]
number3 = input_numbers[num_index[2]]
print('result 2:', number1 * number2 * number3)
```
